## Functions, Conditionals, and Iteration
Let's create a Python function, and call it from a loop.

```
def HelloWorldXY(x, y):
  if (x < 10):
    print("Hello World, x was < 10")
  elif (x < 20):
    print("Hello World, x was >= 10 but < 20")
  else:
    print("Hello World, x was >= 20")
  return x + y

for i in range(8, 25, 5):  # i=8, 13, 18, 23 (start, stop, step)
  print("--- Now running with i: {}".format(i))
  r = HelloWorldXY(i,i)
  print("Result from HelloWorld: {}".format(r))
```

```
print(HelloWorldXY(1,2))
```

```
import os, re, json, random, string
import numpy as np, pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Small helper
def save_csv(df, path):
    Path(path).parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(path, index=False)
    print(f"Saved: {path}")
```

```
# Load dataset
assert os.path.exists(DATA_PATH), f"DATA_PATH not found: {DATA_PATH}"
df = pd.read_csv(DATA_PATH, low_memory=False)
print("Rows:", len(df))
display(df.head())

# Ensure canonical columns exist; rename if necessary
if 'text' not in df.columns:
    # heuristics: find first long text-like column
    for c in df.columns:
        if df[c].astype(str).str.len().mean() > 20:
            df = df.rename(columns={c: 'text'})
            print("Renamed column", c, "-> text")
            break

# Drop rows without text
df = df[~df['text'].isnull()].reset_index(drop=True)
print("After drop empty text:", len(df))
```

```
# Basic shape and label distribution (if label exists)
print("Columns:", df.columns.tolist())
if 'label' in df.columns:
    print("Label distribution:")
    print(df['label'].value_counts(dropna=False))
else:
    print("No 'label' column in data. Please add labels for supervised training (0/1).")

# Quick word count distribution
df['char_len'] = df['text'].astype(str).apply(len)
plt.figure(figsize=(8,4))
sns.histplot(df['char_len'], bins=50)
plt.title("Tweet length distribution (chars)")
plt.show()
```

```
# NLTK downloads & imports
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
sent_analyzer = SentimentIntensityAnalyzer()

# Regex patterns
url_re = re.compile(r'https?://\S+|www\.\S+')
mention_re = re.compile(r'@\w+')
hashtag_re = re.compile(r'#(\w+)')
non_alphanum_re = re.compile(r"[^a-z0-9\s']")

def clean_text(text, lower=True, remove_urls=True, remove_mentions=True):
    if pd.isna(text):
        return ""
    t = str(text)
    if remove_urls:
        t = url_re.sub(" ", t)
    if remove_mentions:
        t = mention_re.sub(" ", t)
    # convert hashtags to words (remove # but keep word)
    t = hashtag_re.sub(r'\1', t)
    if lower:
        t = t.lower()
    t = non_alphanum_re.sub(' ', t)
    t = re.sub(r'\s+', ' ', t).strip()
    return t

def tokenize_lemmatize(text):
    toks = nltk.word_tokenize(text)
    toks = [t for t in toks if t.isalpha() and t not in stop_words]
    toks = [lemmatizer.lemmatize(t) for t in toks]
    return " ".join(toks)

def full_preprocess(text):
    c = clean_text(text)
    return tokenize_lemmatize(c)

# Apply preprocessing to dataset (creates clean_text and sentiment columns)
print("Applying preprocessing (this may take a few seconds)...")
df['clean_text'] = df['text'].astype(str).apply(full_preprocess)
df['sentiment_compound'] = df['text'].astype(str).apply(lambda t: sent_analyzer.polarity_scores(str(t))['compound'])
display(df[['text','clean_text','sentiment_compound']].head(5))
```

```
# If you already have splits.json you can load it, otherwise create stratified split (if labels exist)
if 'label' in df.columns:
    train_idx, test_idx = train_test_split(df.index, test_size=0.2, stratify=df['label'], random_state=SEED)
    val_idx, test_idx = train_test_split(test_idx, test_size=0.5, stratify=df.loc[test_idx,'label'], random_state=SEED)
else:
    # simple random split
    train_idx, rest_idx = train_test_split(df.index, test_size=0.2, random_state=SEED)
    val_idx, test_idx = train_test_split(rest_idx, test_size=0.5, random_state=SEED)

splits = {
    "train_ids": df.loc[train_idx, 'tweet_id'].astype(str).tolist() if 'tweet_id' in df.columns else df.loc[train_idx].index.astype(str).tolist(),
    "val_ids": df.loc[val_idx, 'tweet_id'].astype(str).tolist() if 'tweet_id' in df.columns else df.loc[val_idx].index.astype(str).tolist(),
    "test_ids": df.loc[test_idx, 'tweet_id'].astype(str).tolist() if 'tweet_id' in df.columns else df.loc[test_idx].index.astype(str).tolist()
}

# Pad lists to the same length to allow DataFrame creation
max_len = max(len(splits["train_ids"]), len(splits["val_ids"]), len(splits["test_ids"]))
for key in splits:
    current_list = splits[key]
    if len(current_list) < max_len:
        splits[key] = current_list + [None] * (max_len - len(current_list))

save_csv(pd.DataFrame(splits), "/content/splits_preview.csv")
# Save full df cleaned
save_csv(df, "/content/raw_tweets_cleaned.csv")
print("Train/Val/Test sizes:", len(train_idx), len(val_idx), len(test_idx))
```

```
# Ensure labels exist
if 'label' not in df.columns:
    raise ValueError("No 'label' column present. Add labels (0/1) to train supervised models.")

# Prepare text and labels using the same splits above
X_text = df['clean_text'].fillna("")
y = df['label'].astype(int)

# Fit TF-IDF on training set only
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
tfidf.fit(df.loc[train_idx, 'clean_text'].fillna(""))

X_train = tfidf.transform(df.loc[train_idx, 'clean_text'].fillna(""))
X_val = tfidf.transform(df.loc[val_idx, 'clean_text'].fillna(""))
X_test = tfidf.transform(df.loc[test_idx, 'clean_text'].fillna(""))

# Optionally add sentiment as numeric feature (append to sparse matrix)
from scipy.sparse import hstack
X_train_full = hstack([X_train, np.array(df.loc[train_idx,'sentiment_compound']).reshape(-1,1)])
X_val_full = hstack([X_val, np.array(df.loc[val_idx,'sentiment_compound']).reshape(-1,1)])
X_test_full = hstack([X_test, np.array(df.loc[test_idx,'sentiment_compound']).reshape(-1,1)])

y_train = y[train_idx]
y_val = y[val_idx]
y_test = y[test_idx]

# Train with class weight (help imbalance)
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=SEED)
lr.fit(X_train_full, y_train)

# Evaluate
y_pred = lr.predict(X_test_full)
print("TF-IDF + LR F1:", f1_score(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=4))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d'); plt.title("Confusion matrix - TFIDF+LR"); plt.show()

# Save artifacts
import joblib
joblib.dump(tfidf, "/content/tfidf_vectorizer.joblib")
joblib.dump(lr, "/content/tfidf_lr.joblib")
print("Saved TF-IDF vectorizer and LR model.")
```

```
from sentence_transformers import SentenceTransformer
embed_model = SentenceTransformer('all-MiniLM-L6-v2')  # lightweight & fast
# encode all texts (can take time)
texts = df['text'].astype(str).tolist()
embeddings = embed_model.encode(texts, show_progress_bar=True, batch_size=64)

# Prepare training/test embeddings using splits
X_emb = embeddings
X_emb_train = X_emb[train_idx]
X_emb_val = X_emb[val_idx]
X_emb_test = X_emb[test_idx]

# Optionally append sentiment as a feature
X_emb_train = np.hstack([X_emb_train, df.loc[train_idx,'sentiment_compound'].values.reshape(-1,1)])
X_emb_val = np.hstack([X_emb_val, df.loc[val_idx,'sentiment_compound'].values.reshape(-1,1)])
X_emb_test = np.hstack([X_emb_test, df.loc[test_idx,'sentiment_compound'].values.reshape(-1,1)])

# Train simple classifier
lr_emb = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=SEED)
lr_emb.fit(X_emb_train, y_train)
y_pred_emb = lr_emb.predict(X_emb_test)
print("Embeddings + LR F1:", f1_score(y_test, y_pred_emb))
print(classification_report(y_test, y_pred_emb, digits=4))
joblib.dump(lr_emb, "/content/emb_lr.joblib")
print("Saved embedding classifier.")
```

```
# WARNING: BERT fine-tuning requires GPU. Ensure runtime accelerator is GPU.
import torch
from torch.utils.data import Dataset, DataLoader
# Fix: AdamW is now imported from torch.optim or transformers.optimization
from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup
from torch.optim import AdamW

MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

MAX_LEN = 128
BATCH_SIZE = 16
EPOCHS = 3
LR = 2e-5

class TweetDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
    def __len__(self):
        return len(self.texts)
    def __getitem__(self, idx):
        txt = str(self.texts[idx])
        enc = self.tokenizer(txt, truncation=True, padding='max_length', max_length=self.max_len, return_tensors='pt')
        item = {k: enc[k].squeeze(0) for k in enc}
        item['labels'] = torch.tensor(int(self.labels[idx]), dtype=torch.long)
        return item

train_texts = df.loc[train_idx, 'text'].tolist()
val_texts = df.loc[val_idx, 'text'].tolist()
train_labels = y_train.tolist()
val_labels = y_val.tolist()

train_dataset = TweetDataset(train_texts, train_labels, tokenizer, MAX_LEN)
val_dataset = TweetDataset(val_texts, val_labels, tokenizer, MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)
model = model.to(device)

optimizer = AdamW(model.parameters(), lr=LR)
total_steps = len(train_loader) * EPOCHS
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=int(0.1*total_steps), num_training_steps=total_steps)

def train_epoch(model, loader, optimizer):
    model.train()
    losses = []
    for batch in loader:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()
        losses.append(loss.item())
    return np.mean(losses)

def eval_model(model, loader):
    model.eval()
    preds, labs = [], []
    with torch.no_grad():
        for batch in loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
            labs.extend(labels.cpu().numpy())
    return np.array(preds), np.array(labs)

best_f1 = 0
for epoch in range(EPOCHS):
    tr_loss = train_epoch(model, train_loader, optimizer)
    preds_val, labs_val = eval_model(model, val_loader)
    val_f1 = f1_score(labs_val, preds_val)
    print(f"Epoch {epoch+1} ‚Äî train_loss: {tr_loss:.4f} val_f1: {val_f1:.4f}")
    # save best
    if val_f1 > best_f1:
        best_f1 = val_f1
        model.save_pretrained("/content/bert_purchase_intent")
        tokenizer.save_pretrained("/content/bert_purchase_intent")
        print("Saved best model.")

# Evaluate on test set
test_texts = df.loc[test_idx, 'text'].tolist()
test_labels = y_test.tolist()
test_dataset = TweetDataset(test_texts, test_labels, tokenizer, MAX_LEN)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, num_workers=2, shuffle=False)
preds_test, labs_test = eval_model(model, test_loader)
print("BERT test F1:", f1_score(labs_test, preds_test))
print(classification_report(labs_test, preds_test, digits=4))
```

```
from sklearn.feature_extraction.text import CountVectorizer
def top_ngrams(corpus, n=20, ngram_range=(1,1)):
    vec = CountVectorizer(ngram_range=ngram_range, stop_words='english').fit(corpus)
    bag = vec.transform(corpus)
    sums = bag.sum(axis=0)
    terms = [(term, sums[0, idx]) for term, idx in vec.vocabulary_.items()]
    terms = sorted(terms, key=lambda x: x[1], reverse=True)[:n]
    return terms

if 'label' in df.columns:
    pi_corpus = df[df['label']==1]['clean_text'].astype(str).tolist()
    no_corpus = df[df['label']==0]['clean_text'].astype(str).tolist()
    for name, corpus in [("PI_unigrams", (pi_corpus, (1,1))), ("PI_bigrams", (pi_corpus, (2,2))),
                         ("NoPI_unigrams",(no_corpus,(1,1))), ("NoPI_bigrams",(no_corpus,(2,2)))]:
        terms = top_ngrams(corpus[0], n=25, ngram_range=corpus[1])
        terms_df = pd.DataFrame(terms, columns=['ngram','count'])
        plt.figure(figsize=(8,6))
        sns.barplot(y='ngram', x='count', data=terms_df)
        plt.title(name)
        save_fig(f"{name}.png")
        plt.show()
else:
    print("Top n-grams skipped (no label column).")
```

```
from sklearn.metrics import roc_curve, precision_recall_curve, auc

def plot_pr_roc(y_true, probas, name):
    fpr, tpr, _ = roc_curve(y_true, probas)
    prec, rec, _ = precision_recall_curve(y_true, probas)
    roc_auc = auc(fpr, tpr)
    pr_auc = auc(rec, prec)
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(fpr, tpr, label=f"AUC={roc_auc:.3f}")
    plt.plot([0,1],[0,1],'--', color='grey')
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title(f"ROC - {name}"); plt.legend()
    plt.subplot(1,2,2)
    plt.plot(rec, prec, label=f"PR-AUC={pr_auc:.3f}")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"Precision-Recall - {name}"); plt.legend()
    save_fig(f"{name}_roc_pr.png")
    plt.show()

# TF-IDF LR (probabilities)
try:
    if hasattr(lr, "predict_proba"):
        probs = lr.predict_proba(X_test_full)[:,1]
        plot_pr_roc(y_test, probs, "TFIDF_LR")
    else:
        print("TF-IDF LR has no predict_proba")
except Exception as e:
    print("TF-IDF ROC/PR skipped:", e)

# Embedding LR
try:
    if hasattr(lr_emb, "predict_proba"):
        probs_e = lr_emb.predict_proba(X_emb_test)[:,1]
        plot_pr_roc(y_test, probs_e, "Emb_LR")
except Exception as e:
    print("Embeddings ROC/PR skipped:", e)

# BERT probabilities (softmax)
try:
    import torch, torch.nn.functional as F
    model.eval()
    all_probs = []
    all_labels = []
    for batch in DataLoader(TweetDataset(test_texts, test_labels, tokenizer, MAX_LEN), batch_size=32):
        with torch.no_grad():
            input_ids = batch['input_ids'].to(device)
            att = batch['attention_mask'].to(device)
            out = model(input_ids=input_ids, attention_mask=att)
            probs = F.softmax(out.logits, dim=1)[:,1].cpu().numpy()
            all_probs.extend(probs)
    plot_pr_roc(labs_test, np.array(all_probs), "BERT")
except Exception as e:
    print("BERT ROC/PR skipped:", e)
```

```
def precision_at_k(y_true, scores, ks=[0.001,0.005,0.01,0.02,0.05,0.1]):
    n = len(scores)
    order = np.argsort(scores)[::-1]
    results = []
    for k in ks:
        topk = int(np.ceil(k*n))
        if topk < 1: topk = 1
        prec = y_true[order[:topk]].mean()
        results.append((k, prec))
    return results

# example for TF-IDF LR
try:
    probs = lr.predict_proba(X_test_full)[:,1]
    res = precision_at_k(np.array(y_test), probs, ks=[0.001,0.005,0.01,0.02,0.05,0.1])
    ks = [r[0] for r in res]; precs = [r[1] for r in res]
    plt.figure(figsize=(6,4))
    plt.plot(ks, precs, marker='o')
    plt.xlabel("k (fraction of test set)")
    plt.ylabel("Precision@k")
    plt.title("Precision@k ‚Äî TFIDF LR")
    save_fig("precision_at_k_tfidf.png")
    plt.show()
except Exception as e:
    print("Precision@k skipped:", e)
```

```
from sklearn.calibration import calibration_curve

def plot_calibration(y_true, probs, name):
    fraction_of_positives, mean_predicted_value = calibration_curve(y_true, probs, n_bins=10)
    plt.figure(figsize=(6,6))
    plt.plot(mean_predicted_value, fraction_of_positives, "s-", label=name)
    plt.plot([0,1],[0,1],"--", color='grey')
    plt.xlabel("Mean predicted value")
    plt.ylabel("Fraction of positives")
    plt.title(f"Calibration plot ‚Äî {name}")
    save_fig(f"calibration_{name}.png")
    plt.show()

try:
    if hasattr(lr, "predict_proba"):
        probs = lr.predict_proba(X_test_full)[:,1]
        plot_calibration(y_test, probs, "TFIDF_LR")
    if hasattr(lr_emb, "predict_proba"):
        probs_e = lr_emb.predict_proba(X_emb_test)[:,1]
        plot_calibration(y_test, probs_e, "Emb_LR")
    # BERT
    try:
        plot_calibration(labs_test, np.array(all_probs), "BERT")
    except:
        pass
except Exception as e:
    print("Calibration plots skipped:", e)
```
